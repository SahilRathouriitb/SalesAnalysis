```python
#now our first task will be to import the csv file into our python programme as a panda data frame 
df=pd.read_csv("F:\\pandas\\diwali.csv",encoding='unicode_escape')
#now we have imported the csv file in our programme 
#we cant print it since it has a lot of rows and columns
```

# GENDER DISTRIBUTION IN THE SURVEY

```python
#now the very first one will be on the basis of gender that is we will see the divison of gender ...
ax=sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
    #here overall we have counted the number of males and females and we are pretty much happy

```

```python
# WE CAN CONCLUDE THAT IS PRODUCT BASED COMPANY HAS MORE FEMALE BUYERS AND THEIR PRODUCT IS PREFERRED MORE BY FEMALES
```

# GENDER VS AMOUNT SPENT

```python
#now lets say i want to see what gender has spent what aount of money usig data visualisation
#so we will have to group by gender with amount 
#that is we want to have what gender is spending what amount 
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax2=sns.barplot(x='Gender',y='Amount',data=sales_gen)
for bars in ax2.containers:
     ax2.bar_label(bars)
#so here we have plotted gender versus amount graph using group by from pandas and barplot from seaborn
```

```python
# FROM THIS GRAPH WE CAN SEE THAT FRMALES HAVE SPENT MORE AMOUNT WHICH IS OBVIOUS SINCE FEMALES DID MORE PURCHASES
```

# AGE GROUP AND PEOPLE

```python
#now we will plot a countplot for age i.e how many people are from which age group
ax=sns.countplot(x='Age Group',data=df,hue='Gender')
#now we will label the bars
for bar in ax.containers:
    ax.bar_label(bar)
    
```

```python
# FROM THIS WE CAN SAY THAT THE PRODUCTS ARE BEING LIKED BY FEMALES AND MALES OAGED BTWEEN 26-30 AND EVEN BY OLD PEOPLE , WHERE AS YOUNF TEENS DONT PREFER THE PRODUCTS SOLD BY THIS COMPANY
```

# STATE VS AMOUNT SPENT

# OCCUPATION VS AMOUNT
